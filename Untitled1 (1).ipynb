```python
import pandas as pd
import numpy as sp
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv("housing.csv")# delimeter was not specified as it was already given
#median house value is the target attribute
#Our data first needs to be preprocessed- we can start at the attribute-Ocean proximity

```

```python
data.info()#before that ,the non_null numbers must all be the same
```

```python
#to solve this we do the following:
data.dropna(inplace = True) #we dropped non null values and so out dataframe  we only display non null,to add this saved data use(inplace=true)
```

```python
data.info()
```

```python
#next we have to split the data into training and testing data and into x and y, we will work with only a piece of the data to see how the model responds
```

```python
from sklearn.model_selection import train_test_split
x = data.drop(['median_house_value'],axis=1)# x and y need to be defined first
y = data['median_house_value']
```

```python
x
```

```python
y
```

```python
x_train,x_test,y_train,y_test = train_test_split(x, y, test_size = 0.2)#this data(the test set) will not be touched until we are confident we can launch our model
```

```python
train_data = x_train.join(y_train)#this joins the once split x and y data to see if we can identify any correlations
```

```python
train_data
```

```python
#exploration of some numeric features
train_data.hist(figsize=(15,8))#shows histograms of the data (except ocean proximity)
```

```python

train_data.corr(numeric_only=True)
```

```python

plt.figure(figsize=(15,8)) 
sns.heatmap(train_data.corr(numeric_only=True), annot=True)
```

```python
#looking at our histograms we can see that total_rooms,total_bedrooms,population and households are rather skewed
train_data['total_rooms'] = sp.log(train_data['total_rooms'] + 1 )
train_data['total_bedrooms'] = sp.log(train_data['total_bedrooms'] + 1)
train_data['population'] = sp.log(train_data['population'] + 1)
train_data['households'] = sp.log(train_data['households'] + 1)
```

```python
train_data.hist(figsize=(15,8))
```

```python
#now we want to change the "ocean proximity" attribute into a binary
train_data.ocean_proximity.value_counts()
```

```python
#now we want to change the "ocean proximity" attribute into a binary
df = pd.get_dummies(train_data.ocean_proximity)
print(df)
df = df.replace({True:1,False:0})
print(df)
```

```python
#we are going to create a "yes" or "no" for each attribute that makes up ocean proximity
train_data=train_data.join(pd.getdummies(train_data.ocean_proximity).drop(['ocean proximity'], axis = 1))
```
